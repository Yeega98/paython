```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import pandas as pd
```

```python
x=np.linspace(0,5,11)
y=x**2
```

## functional method of visualising data in python using matplotlib

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('chart1')
```

```python
plt.subplot(1,2,1)#nrow,ncol,plot_number
plt.plot(x,y,"r")
plt.subplot(1,2,2)
plt.plot(x,y,"b")
```

## Object oriented method

```python
fig =plt.figure(figsize=(12,4))
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(1,3,figsize=(12,5))
axes[0].plot(x,y,"y")
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('chart1')

axes[1].plot(x,y,"r")
axes[2].plot(x,y,"b")
```

```python
z=[1,2,3,4]
label=["fair","good","very good","excellent"]
```

```python
plt.figure(figsize=(16,8))
plt.bar(label,z)
plt.xlabel('category')
plt.ylabel('value')
plt.title('Bar_chart')
```

```python
plt.figure(figsize=(16,8))
plt.barh(label,z)
plt.ylabel("category")
plt.xlabel("values")
plt.title("horizontal_bar_chart")
```

```python
plt.figure(figsize=(12,5))
plt.pie(z,labels=label);
```
